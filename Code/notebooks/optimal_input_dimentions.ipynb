# Find optimal input image size

```python
import os
import statistics
import matplotlib.pyplot as plt
from PIL import Image
```

```python
dataset_dir = '/Path/To/Base/Dataset/Directory/'
```

```python
heights = []
widths  = []

# Iterate over folders in the source directory
for folder_name in os.listdir(dataset_dir):
    folder_path = os.path.join(dataset_dir, folder_name)

    # Get the initial image
    image_file_path = os.path.join(folder_path, f"{folder_name}_center_0.jpg")

    # Open image
    orig_img = Image.open(image_file_path)

    # Comput height, width
    width, height = orig_img.size

    heights.append(height)
    widths .append(width)
```

## Optimal Height

```python
mean_height = statistics.mean(heights)
print(f'mean value: {mean_height}')

# Create a histogram for the heights
plt.hist(heights, bins=100, alpha=0.5, label='heights', log=True)

# Plot a horizontal line at mean value
plt.axvline(x=mean_height, color='red', linestyle='--', label='mean height')

# Add labels and a legend
plt.xlabel('Values (pixels)')
plt.ylabel('Frequency (log)')
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

## Optimal Width

```python
mean_width = statistics.mean(widths)
print(f'mean value: {mean_width}')

# Create a histogram for the widths
plt.hist(widths, bins=100, alpha=0.5, label='widths', log=True)

# Plot a horizontal line at mean value
plt.axvline(x=mean_width, color='red', linestyle='--', label='mean width')

# Add labels and a legend
plt.xlabel('Values (pixels)')
plt.ylabel('Frequency (log)')
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

## Optimal Aspect Ratio

```python
aspect_ratio = []

for h,w in zip(heights, widths):
    aspect_ratio.append(h/w)
```

```python
mean_aspect_ratio = statistics.mean(aspect_ratio)
print(f'mean value: {mean_aspect_ratio}')

# Create a histogram for the widths
plt.hist(aspect_ratio, bins=100, alpha=0.5, label='aspect ratio', log=True)

# Plot a horizontal line at mean value
plt.axvline(x=mean_aspect_ratio, color='red', linestyle='--', label='mean aspect ratio')

# Add labels and a legend
plt.xlabel('Values (pixels)')
plt.ylabel('Frequency (log)')
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
plt.show()
print(f'Optimal value: {mean_aspect_ratio}')
print()
print('height:width = aspect ratio')
print(f'   352:256   = {352 / 256:.3f},        dev. from optimal: {abs((352 / 256) - mean_aspect_ratio):.3f}')
print(f'   320:224   = {320 / 224:.3f},        dev. from optimal: {abs((352 / 224) - mean_aspect_ratio):.3f}')
print(f'   288:224   = {288 / 224:.3f},        dev. from optimal: {abs((288 / 224) - mean_aspect_ratio):.3f}')
print(f'   256:192   = {256 / 192:.3f},        dev. from optimal: {abs((256 / 192) - mean_aspect_ratio):.3f}')
print(f'   224:160   = {224 / 160:.3f},        dev. from optimal: {abs((224 / 160) - mean_aspect_ratio):.3f}')
print(f'   192:128   = {192 / 128:.3f},        dev. from optimal: {abs((192 / 128) - mean_aspect_ratio):.3f}')
print(f'   160:128   = {160 / 128:.3f},        dev. from optimal: {abs((160 / 128) - mean_aspect_ratio):.3f}')
print(f'   128:96    = {128 / 96 :.3f},        dev. from optimal: {abs((128 / 96) - mean_aspect_ratio):.3f}')
```
